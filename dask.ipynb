Step 1: Import and Read Data

```python
import dask.dataframe as dd
df = dd.read_csv('superstore.csv', dtype={'Discount': 'float64'})
 
df
```

Step 2: Inspect the Data

```python

print(df.dtypes)
```

```python
df.head()
```

Step 3: Common Big Data Operations
1. Filter: Sales over a certain amount

```python

high_value_sales = df[df['Sales'] > 1000]
print(df.columns)

```

2. GroupBy and Aggregate

```python
sales_by_region = df.groupby('Country')['Sales'].sum().compute()
print(sales_by_region)
```

3. Top-selling products

```python
top_products = df.groupby('Sub.Category')['Sales'].sum().nlargest(10).compute()
print(top_products)
```

Visualization 

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
sales_by_region.plot(kind='line', title=' Sales region')
plt.xlabel('Sales')
plt.ylabel('Market')
plt.show()
```

```python
plt.figure(figsize=(8, 8))  
top_products.plot(
    kind='pie',
    autopct='%1.1f%%',       
    startangle=90,           
    legend=False,
    shadow=True              
)

 
plt.title('high sale product')
plt.ylabel('')   
plt.tight_layout()
plt.show()
```
